```python
import csv
import ast

loadexercises = open('/filepath/exercise_dictionary_001-589_revised200428.csv', mode='r', newline = '')
reader = csv.reader(loadexercises)
e_transposable = {}
for row in reader:
    key, value = row
    e_transposable[key] = ast.literal_eval(value)
loadexercises.close()

loadexercises = open('/filepath/exercise_dictionary_431and590.csv', mode='r', newline = '')
reader = csv.reader(loadexercises)
e_nontransposable = {}
for row in reader:
    key, value = row
    e_nontransposable[key] = ast.literal_eval(value)
loadexercises.close()
```

```python
def split_XY(exercise):
  ex_X = []
  ex_Y = []
  for note in exercise:
    ex_X.append(list([note[0], note[2]]))
    ex_Y.append(note[1])

  return(ex_X, ex_Y)

dataset_e_transposable_X = []
dataset_e_transposable_Y = []

for key in e_transposable:
  ex_X, ex_Y = split_XY(e_transposable[key])
  dataset_e_transposable_X.append(list(ex_X))
  dataset_e_transposable_Y.append(list(ex_Y))

dataset_e_nontransposable_X = []
dataset_e_nontransposable_Y = []

for key in e_nontransposable:
  ex_X, ex_Y = split_XY(e_nontransposable[key])
  dataset_e_nontransposable_X.append(list(ex_X))
  dataset_e_nontransposable_Y.append(list(ex_Y))
```

```python
#Detecting errors in the original .xml an array encoding to be fixed by hand.
#This included incorrect pitch numbers and single notes being assigned multiple sticks.

bad_pitch_code = []
good_pitch_code = []
maybe_pitch_code = []

for count, exercise in enumerate(dataset_e_transposable_X):
  pitch_seq = []
  for note in exercise:
    if type(note[0]) == int:
      pitch_seq.append(note[0])
    elif type(note[0]) == list:
      for i in note[0]:
        pitch_seq.append(i)
    else:
      print("Type error")
      break
  min_pitch = min(pitch_seq)
  max_pitch = max(pitch_seq)
  if min_pitch < 24:
    bad_pitch_code.append(count)
  elif max_pitch > 60:
    good_pitch_code.append(count)
  else:
    maybe_pitch_code.append(count)

print(bad_pitch_code)
print(maybe_pitch_code)
```
